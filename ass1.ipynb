```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 8))
top5_brands = data['Brand'].value_counts().head(5)

plt.pie(
    top5_brands.values,
    labels=top5_brands.index,
    autopct='%1.1f%%',
    colors=sns.color_palette('pastel'),
    startangle=140
)

plt.title('Top 5 Brands Distribution')
plt.show() 
plt.savefig("pie.png")
```

```python
import matplotlib.pyplot as plt

top5_brands = data['Brand'].value_counts().head(5)

plt.figure(figsize=(8, 6))
plt.bar(top5_brands.index, top5_brands.values, color='skyblue', edgecolor='black')

plt.title('Top 5 Brands Distribution')
plt.xlabel('Brand')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
plt.savefig("bargraph.png")
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(8, 6))
plt.hist(data['Price'], bins=20, color='skyblue', edgecolor='black')

plt.title('Car Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
plt.savefig("Histogram.png")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))

# Scatter plot with numeric X and Y
sns.scatterplot(x='Mileage', y='Price', data=data, color='skyblue', edgecolor='black', alpha=0.6)

# Labels and title
plt.title('Scatter Plot: Price vs Mileage')
plt.xlabel('Mileage (km/l)')
plt.ylabel('Price')
plt.grid(True)
plt.tight_layout()
plt.savefig("scatter.png")
plt.show()
```
